```python
model = YOLO("best_v13.pt")
```

```python
image1 = "indoor-objects-recognition-13/train/images/2025-11-30_17_36_51-190_jpg.rf.17a20474cbb1c87051bcac58864d5354.jpg"
results = model.predict(image1)
```

```python
# Process results list
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs
    obb = result.obb  # Oriented boxes object for OBB outputs
    result.show()  # display to screen
    result.save(filename="result.jpg")  # save to disk
```

```python
#train model
dataset_dir = "dataset_v14"
path_train = os.path.join(dataset_dir, "train/images")
path_valid = os.path.join(dataset_dir, "valid/images")
path_test  = os.path.join(dataset_dir, "test/images")
yaml_path = os.path.join(dataset_dir, "data.yaml")
```

```python
model.train(
        data=yaml_path,   
        epochs=200,
        imgsz=640,
        batch=24,
        device=0
    )
```
